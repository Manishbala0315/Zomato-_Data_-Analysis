```python
df = df1.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace= True)
df.info()
```

```python
df.rename(columns = {'approx_cost(for two people)':'cost2plates','listed_in(type)':'type' }, inplace = True)
df.info()
```

```python
df.head()
```

```python
df['location'].unique()
```

```python
df = df.drop(['listed_in(city)'] , axis = 1)
```

```python
df.head(3)
```

```python
df['cost2plates'].unique()
```

```python

print("Original DataFrame:")
print(df)


df['cost2plates'] = df['cost2plates'].str.replace(',', '', regex=True)


df['cost2plates'] = pd.to_numeric(df['cost2plates'], errors='coerce')  # Use 'coerce' to handle any conversion issues


print("\nUpdated DataFrame:")
print(df)
```

```python
df['cost2plates'].unique()
```

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_type = df['rest_type'].value_counts(ascending= False)
```

```python
rest_type_less1000 = rest_type[rest_type <1000]
rest_type_less1000
```

```python
def handLe_rest(value):
  if(value in rest_type_less1000):
    return 'othors'
  else:
    return value


df['rest_type'] = df['rest_type'].apply(handLe_rest)
df['rest_type'].value_counts()
```

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False )
location_less300 = location[location<300]
```

```python
def handle_location(value):
  if(value in location_less300):
    return 'othors'
  else:
    return value


df['location']= df['location'].apply(handle_location)
df["location"].value_counts()
```

```python
df.head()
```

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False )

cuisines_less100 = cuisines[ cuisines<100]
```

```python
def handle_cuisines(value):
  if( value in cuisines_less100):
    return "othors"
  else:
    return value

df['cuisines'] = df["cuisines"].apply(handle_cuisines)
df['cuisines'].value_counts()
```
